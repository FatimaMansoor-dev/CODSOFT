```python
## 0s are not possible in time, this seems like an error, there must have been atleast 15 seconds gap between two transactions
```

```python
# lets check how many values in amount have 0
df[df.Amount == 0].count()[0]
```

```python
# transaction amount=0 makes no sense, they are null values or some error
df["Amount"] = df["Amount"].replace(0,np.NaN)
df["Time"] = df["Time"].replace(0,np.NaN)
```

```python
df.isnull().sum()
```

```python
df2 = df.dropna(subset=['Time', 'Amount'])
df2.head()
```

## smote for imbalance

```python
x= df3.drop("Class",axis = 1)
y = df3.Class
```

```python
x
```

```python
y
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)
x_sm,y_sm = sm.fit_resample(x,y)
```

```python
corr_bal = x_sm.corr()
sns.heatmap(corr_bal,annot=False,cmap="plasma",linewidth = 0.5)
plt.title("balanced data")
```

```python
sns.countplot(x=y_sm)
plt.title("Distribution of classes")
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_sm,y_sm)
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(x_train,y_train)
```

```python
train_pred = reg.predict(x_train)
test_pred = reg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print("train acc : ",accuracy_score(train_pred,y_train))
print("test acc : ",accuracy_score(test_pred,y_test))
```

# random forest

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier  
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score  
```

```python
model = RandomForestClassifier(n_estimators=100)  # Number of trees in the forest
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
accuracy = accuracy_score(y_test, predictions)
```

```python
accuracy
```

```python
## train accuracy
train_pred2 = model.predict(x_train)
print("train acc : ",accuracy_score(train_pred2,y_train))
```
